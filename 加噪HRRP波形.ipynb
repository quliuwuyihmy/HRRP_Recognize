```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import math
import matplotlib.pyplot as plt
import numpy as np


img = np.zeros((1200,256),dtype=np.float64)
i = 0
for line in open("/Users/yuchen/Desktop/HRR/hrr.dat",'r'):
    img[i] = line.split() 
    i += 1
    
    

pure_signal = [img[100],img[302],img[702],img[1010]]
```
